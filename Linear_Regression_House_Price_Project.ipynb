```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_california_housing

# Load the dataset (using fetch_california_housing as the example data)
data = fetch_california_housing()

# Create a DataFrame for easier manipulation
df = pd.DataFrame(data['data'], columns=data['feature_names'])

# Generate a box plot for each feature
plt.figure(figsize=(15, 10))
sns.boxplot(data=df)
plt.xticks(rotation=90)
plt.title("Box Plot of Each Feature")
plt.show()
```

```python
dataset.head()
```

```python
#Remove population as they are outliers
# Drop a specific feature 'population'
feature_to_drop = 'Population'
df_dropped = df.drop(columns=[feature_to_drop])

```

```python
# Display the DataFrame after dropping the feature
print("\nDataFrame after dropping the feature:")
print(df_dropped.head())
```

```python
## Independent features and dependent features
X=df.drop('MedInc', axis = 1)
y=df.MedInc
```

```python
y
```

```python
## train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.30, random_state=42)
```

```python
X_train
```

```python
## Standarizing the dataset
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
## cross validation (Cross-validation is a technique used in machine learning to assess how well a model will generalize to an independent dataset.Overall, cross-validation is a critical step in building robust machine learning models, as it helps ensure that the model performs well not just on the training data but also on unseen data.)
from sklearn.model_selection import cross_val_score
```

```python
regression=LinearRegression()
regression.fit(X_train,y_train)
```

```python
mse=cross_val_score(regression,X,y,scoring='neg_mean_squared_error',cv=5)
```

```python
np.mean(mse)
```

```python
# Calculate RMSE
# Negate the MSE values to get positive MSE
mse = -mse

# Calculate RMSE
rmse = np.sqrt(mse.mean())

print("RMSE:", rmse)
```

```python
## prediction
reg_predict=regression.predict(X_test)
```

```python
reg_predict
```

```python
import seaborn as sns
sns.displot(reg_predict-y_test,kind='kde')
```

```python
##here in above graph we have good graph and our variance is less so it is good.
```

```python
from sklearn.metrics import r2_score
```

```python
score=r2_score(reg_predict,y_test)
```
